### 定義標準化函式

```python
def Std(X, nf):
    X_std = np.copy(X)
    for i in range(nf):
        X_std[:, i] = (X_std[:, i] - X[:, i].mean())/ X_std[:, i].std()
    return X_std
```

## 2.定義演算法函式(Perceptron、Adaline、AdalineSGD)

### (1) Perceptron 

> eta:　learning rate，預設0.01<br>
> n_inter:　epoch number<br>
> random_state:　隨機變數的因子<br>
> fit(): training model, 隨機產生一組weight，把input丟進去train n_iter次，update參數<br>
> net_input(): 根據input X陣列，算矩陣dot的結果(一個常數, wx)<br>
> predict(): 根據net_input出來的值(是否大於0，treshhold function)，output 1 or -1(1為true)<br>
> errors_: 紀錄每次跑完的error

```python
class Perceptron(object):
    def __init__(self, eta=0.01, n_iter=10, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        
    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.errors_ = []
        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

### (2) Adaline

> 和Perceptron大部分都一樣，主要不一樣在fit的過程<br>
  w的變量會根據 training rate 跟 errors決定，learning rate tune得不要太大會慢慢收斂<br>
  cost_是拿來紀錄每次跑完一個epoch的cost

```python
class Adaline(object):
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            output = self.activation(X)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        return self.net_input(X)

    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

### (3) AdalineSGD

> _update_weights(): 拉出來定義，會吐一個cost給fit()紀錄
> fit():初始化w(用_intialize_weights，small number)，資料是一筆一筆送進去update(用zip那一行)，速度快<br>
> partial_fit(): fit data但是用tune過的參數，可以在之後送新的data的時候繼續train

```python
class AdalineSGD(object):
    def __init__(self, eta=0.01, n_iter=50, shuffle=True, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.w_initialized = False
        self.shuffle = shuffle
        if random_state:
            seed(random_state)
        
    def fit(self, X, y):
        self._initialize_weights(X.shape[1])
        self.cost_ = []
        for i in range(self.n_iter):
            if self.shuffle:
                X, y = self._shuffle(X, y)
            cost = []
            for xi, target in zip(X, y):
                cost.append(self._update_weights(xi, target))
            avg_cost = sum(cost) / len(y)
            self.cost_.append(avg_cost)
        return self

    def _shuffle(self, X, y):
        r = np.random.permutation(len(y))
        return X[r], y[r]
    
    def _initialize_weights(self, m):
        self.w_ = np.random.normal(loc=0.0, scale=0.01, size=1 + m)
        self.w_initialized = True
        
    def _update_weights(self, xi, target):
        output = self.activation(xi)
        error = (target - output)
        self.w_[1:] += self.eta * xi.dot(error)
        self.w_[0] += self.eta * error
        cost = 0.5 * error**2
        return cost
    
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        return self.net_input(X)

    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)

    def partial_fit(self, X, y):
        if not self.w_initialized:
            self._initialize_weights(X.shape[1])
        if y.ravel().shape[0] > 1:
            for xi, target in zip(X, y):
                self._update_weights(xi, target)
        else:
            self._update_weights(X, y)
        return self
```

### 定義OvR分類器物件

為了讓code簡潔，特別在寫一個物件去執行OvR：
* init : 宣告時輸入eta, n_iter, 作為內部每個分類器的參數
         n_class是class的種類
         algo是要選用的演算法，用eval函式可以根據這個字串產生不同的module
         宣告一個list去存n_class個module
* fit : 輸入訓練集，一次訓練3個module
* predict : 輸入測試集，比較每一個model都出的net_input大小，選擇較大的分類器輸出預測，並輸出list

```python
class OvR(object):
    def __init__(self, eta=0.01, n_iter=10, n_class = 3, algo = 'Perceptron', random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        self.n_class = n_class
        self.ppn_list = []
        for i in range(n_class):
            self.ppn_list.append(eval(algo + "(eta = self.eta, n_iter = self.n_iter)"))
            
    def fit(self, X, Y):
        j = 0
        for i in self.ppn_list:
            y = np.where(Y == j, 1, -1)
            i.fit(X , y)
            j += 1
            
    def predict(self, X):
        z_list = []
        y_list = []
        for i in self.ppn_list:
            z_list.append(i.net_input(X))
        z = np.dstack(z_list)
        z = np.reshape(z, (len(X), self.n_class))
        for l in z:
            l = l.tolist()
            y_list.append(l.index(max(l)))
        return y_list
```

### 定義準確率函式

```python
def ACC(pre, test):
    a = zip(pre, test)
    hit = 0
    for i in a:
        if i[0] == i[1]:
            hit += 1
    return hit / len(pre)
```
